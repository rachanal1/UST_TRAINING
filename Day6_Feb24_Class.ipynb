```python
#Histogram
```

```python
import matplotlib.pyplot as plt
import numpy as np

data=np.random.random_sample(5) #genrating random data for histogram

plt.hist(data,color='green',edgecolor='yellow',bins=5) #5 equal widths

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram plot')
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,3,5,7,9]
y=[12,14,16,18,20]

plt.scatter(x,y,color='blue',marker='*')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('scatter plot')
plt.show()
```

```python
#pie chart
```

```python
categories=['shopping','food','travel','savings','health']
expenses=[800,400,500,2000,1000]

plt.figure(figsize=(5,10))
plt.pie(expenses,labels=categories,autopct='%2.2f%%')
plt.legend()
plt.savefig('piechart.png')
plt.show()
```

```python
#stack plot (stacked area plot)
```

```python
import matplotlib.pyplot as plt

quarters=['Q1','Q2','Q3','Q4']
electronics=[14000,13000,15000,10500]
clothing=[4000,6000,7500,8000]
home_appliances=[3000,7500,8200,9000]

plt.figure(figsize=(10,6)) #width height of figure
plt.stackplot(quarters,electronics,clothing,home_appliances,labels=['Electronics','clothing','home appliances'])

plt.xlabel('quarters')
plt.ylabel('sales ($)')
plt.title('product category sales over quarters')
#plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
x,y=np.meshgrid(x,y)
z=np.sin(np.sqrt(x**2 + y**2))

fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x,y,z,c=z,cmap='viridis')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

```python
#sub plot

import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,15,75)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
y4=np.exp(-x)
y5=np.sqrt(x)

fig,axs=plt.subplots(2,3,figsize=(8,8))
axs[0,0].plot(x,y1,'red')
axs[0,0].set_title('sine wave')
axs[0,1].plot(x,y2,'green')
axs[0,1].set_title('cosine wave')
axs[1,0].plot(x,y3,'blue')
axs[1,0].set_title('tangent wave')
axs[1,1].plot(x,y4,'orange')
axs[1,1].set_title('Exponential decay')
axs[0,2].plot(x,y5,'brown')
axs[0,2].set_title('sqrt')

plt.tight_layout() #adjust layout
plt.show()
```
