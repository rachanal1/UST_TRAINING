```python
#Assessment hcf of numbers
def hcf(a, b):
    while b != 0:
        a, b = b, a % b
    return a

n=hcf(36, 60)
print(n)
```

```python
#factors of number
def find_fact(n):
    factors=[]
    for i in range(1,n+1):
        if n%i==0:
            factors.append(i)
    return factors
n1=find_fact(15)
print("Factors are: ",n1)
```

```python
#reduce is for performing some computation on list and returns the result
#it applies rolling computation to sequential pairs of values in list

from functools import reduce
lst=[2,4,5,8]
def mul(x,y):
    return x*y
product=reduce(mul,lst)
#print(product)


product=reduce(lambda x,y:x*y,lst)
print(product)
```

```python
#Inheritance
class sports():
    def __init__(self,model,color,cost):
        self.model=model
        self.color=color
        self.cost=cost
    def getmodel(self):
        return self.model
    def setmodel(self,name):
        self.model=name
    def getcolor(self):
        return self.color
    def setmodel(self,color):
        self.color=color
    def getcost(self):
        return self.cost
    def setcost(self,cost):
        self.cost=cost
class shoes(sports):
    def __init__(self,model,color,cost,brand):
        super().__init__(model,color,cost)
        self.brand=brand
    def shoeinfo(self):
        return f"Model: {self.model}, Color: {self.color}, Cost: {self.cost}, Brand: {self.brand}"

sh1=shoes("shoes","white","Rs.1000","xyz")
print(sh1.shoeinfo())
print(sh1.getmodel())
    
```
