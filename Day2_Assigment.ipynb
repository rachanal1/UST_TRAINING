```python
#where to use map,reduce and filter
```

```python
#Map
#we use map like when we want to apply the condition to all elements
lst=[1,3,9,29]
lst2=list(map(lambda x:x*5,lst))
print(lst2)
```

```python
#Filter
#here filter keeps only those that matches the condition
#it removes unwanted items

nums=[1,3,6,4,7,9,12]
evens=list(filter(lambda x:x%2==0,nums))
print(evens)
```

```python
#Reduce
#it combines everythung into one value
#we use this when we want to collapse list into single result

from functools import reduce
lst=[1,3,11,18]
sum=reduce(lambda x,y:x+y,lst)
print(sum)
```

```python
#why cant we use local variables outside the function
```

```python
#Local variables are created inside a function and exists while function is running
#they are private to the function
#they are isolated so we can reuse the function safely
#locals are created while running the function and it destroys when function finishes but if we use them globally it takes more memory usage, so if we use locally python cleans them automatically
#they are not used outside function because function are to be safe and memory efficient
#if we use globally it causes name conflict
```

```python
tax=0.1
def calculate_total(price):
    global tax
    total=price+price*tax
    return total

def calculate_discount(price):
    global tax
    tax=0.2
    discount=price-price*tax
    return discount


print(calculate_total(100))
print(calculate_discount(100))
print(calculate_total(100))


#here in calculate_discount tax is 0.2 and calculate accordingly and for calaculate_total tax is 0.2 because calculate_discount chnaged it
#so it causes name conflict
#we use local to avoid name conflicts

     
```

```python
#below one is right way to use by using local variable

def calculate_total(price):
    tax=0.1
    total=price+price*tax
    return total

def calculate_discount(price):
    tax=0.2
    discount=price-price*tax
    return discount


print(calculate_total(100))
print(calculate_discount(100))
print(calculate_total(100)) 
```
