```python
#creating directory

import os

os.mkdir('Day3assign')
```

```python
#putting file

filepath=os.path.join('Day3assign','ass.txt')
```

```python
#writing data

lst=open(filepath,'w')
lst.write("This is assignment of day3 \n")
lst.write("we discussed about python file directory")
lst.close()
```

```python
#reading data

lst=open(filepath,'r')
content=lst.read()
print(content)
```

```python
lst=open(filepath,'r')
lst.readline() #reading line by line
```

```python
lst.readline() #reading line by line
```

```python
lst=open(filepath,'r')
lst.readlines() #reads entire content at a time
```

```python
#we can use for lopp also to read line by line
lst=open(filepath,'r')
for line in lst:
    print(line)
```

```python
os.remove(filepath)
```
