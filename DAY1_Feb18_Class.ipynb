```python
#prime number
for num in range(2, 11):
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num)
```

```python
#to check num is prime or not
num = 5
is_prime = True
for i in range(2, num):
    if num % i == 0:
        is_prime = False
        break

if is_prime:
    print("yes") 
else:
    print("no")  
```

```python
lst={1,2,3,4,5}
lst2={5,6,7,8}
print(lst | lst2)
print(lst.union(lst2))
print(lst & lst2)
print(lst.intersection(lst2))
print(lst - lst2)
print(lst.difference(lst2)) #here order matters it removes common elements and gives the remaining elements from lst
```

```python
#symmetric operator
#use ^ this operator

print(lst ^ lst2)
print(lst.symmetric_difference(lst2)) #elements from both sets but it exclude which are in common
```
