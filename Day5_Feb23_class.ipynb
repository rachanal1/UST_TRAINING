```python
def set_season(EST):
    month=EST.month
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Fail"
```

```python
df['EST']=pd.to_datetime(df['EST'])
df['season']=df['EST'].apply(set_season)
```

```python
df['season'].value_counts()
```

```python
print(df['EST'][df['Humidity']==70])
print(df['EST'][df['Temperature']>=70])
print(df['EST'][df['WindSpeedMPH']==7.0])
```

```python
import matplotlib.pyplot as plt
months=['jan','feb','mar','apr','may','jun']
product_a_revenue=[45,30,50,40,65,78]
product_b_revenue=[35,40,50,55,70,68]

plt.plot(months,product_a_revenue,marker='*',linestyle='-',color='blue',label='product A')
plt.plot(months,product_b_revenue,marker='o',linestyle='-',color='red',label='product B')

plt.xlabel('Month')
plt.ylabel('Monthly Revenue (in $1000)')
plt.title('Monthly Revenue comparison')

plt.legend() #display legend to differentiate between product A and product B
plt.show()
```

```python
import pandas as pd
df=pd.read_csv('IPL Matches 2008-2020.csv')
df
```

```python
df.describe()
```

```python
df.shape
```

```python
df['winner'].value_counts()
```

```python
df['season']=pd.to_datetime(df['date']).dt.year
df['toss_decision_num']=df['toss_decision'].map({'bat':0,'field':1})
plt.figure()
plt.scatter(df['toss_decision_num'],df['result_margin'],color='blue')
plt.xlabel('Toss decision')
plt.ylabel('Result margin')
plt.title('Toss decision vs Match Result margin')
plt.show()
```

```python
plt.figure()
plt.hist(df['result_margin'].dropna(),bins=20)
plt.xlabel('Result_margin')
plt.ylabel('frequency')
plt.title('Distribution of match result margin')
plt.show()
```

```python
result_type=df['result'].value_counts()
plt.figure()
result_type.plot(kind='pie',autopct='%1.1f%%')
plt.ylabel("")
plt.title("Match result type (tie,normal,no resut)")
plt.show()
```

```python
team_season_wins=(df.groupby(['season','winner']).size().unstack(fill_value=0))
plt.figure()
plt.imshow(team_season_wins,aspect='auto')
plt.yticks(range(len(team_season_wins.index)),team_season_wins.index)
plt.xticks(range(len(team_season_wins.columns)),team_season_wins.columns,rotation=90)
plt.xlabel('Team')
plt.ylabel('season')
plt.title('Team wins by season')
plt.colorbar(label='wins')
plt.show()
```

```python
win_by_teams=df['winner'].value_counts().head(10)
plt.figure()
win_by_teams.plot(kind='barh',color=['red','blue','green','pink','skyblue','orange','brown'])
plt.xlabel('Team')
plt.ylabel('wins')
plt.title('Wins by Top 10 teams')
plt.show()
```
