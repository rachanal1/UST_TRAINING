```python
iterator and generator
iterator allows to traverse through all the elements of collection(list and tuple)one at a time
implements __iter__() and __next__() methd=ods
used in for loop
```

```python
class Mynum:
    def __iter__(self):
        self.num=1
        return self
    def __next__(self):
        if self.num<=5:
            x=self.num
            self.num+=1
            return x
        else:
            raise StopIteration
mycls=Mynum()
myitr=iter(mycls)

for x in myitr:
    print(x)
```

```python
class num:
    def __iter__(self):
        self.n=1
        return self
    def __next__(self):
        if self.n<=50:
            x=self.n
            self.n=self.n**2+1
            return x
        else:
            raise StopIteration
n1=num()
n2=iter(n1)

for x in n2:
    print(x)
```

```python
generator a speial type of iterator that is defined using function and 'yield' keyword
uses 'yield' instead of 'return'
generate values lazily(like one at a time)
more memory efficient than normal functions
```

```python
def my_gen():
    for i in range(1,6):
        yield i

gen=my_gen()

for value in gen:
    print(value)
```

```python
def my_generator():
    for i in range(1,50):
        i=i**2-1
        yield i

gen=my_generator()

for val in gen:
    print(val)
```

```python
iterator vs generator

iterator: requires __iter__() and __next__()
generator: uses function with yield keyword
iterator: more code and manual implementation
generator: simple and concise implementation
iterator: slightly more control
generator: more memory efficient
```

```python
# Iterator for even numbers
class EvenNumbers:
    def __init__(self, limit):
        if not isinstance(limit, int) or limit < 0:
            raise ValueError("Limit must be a non-negative integer.")
        self.limit = limit
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.limit:
            raise StopIteration
        value = self.current
        self.current += 2
        return value


# Generator for odd numbers
def odd_numbers(limit):
    if not isinstance(limit, int) or limit < 0:
        raise ValueError("Limit must be a non-negative integer.")
    num = 1
    while num <= limit:
        yield num
        num += 2
        
if __name__ == "__main__":
    try:
        limit = int(input("Enter a non-negative integer limit: "))

        print("\nEven numbers using Iterator:")
        even_iter = EvenNumbers(limit)
        for num in even_iter:
            print(num, end=" ")

        print("\n\nOdd numbers using Generator:")
        for num in odd_numbers(limit):
            print(num, end=" ")

    except ValueError as e:
        print(f"Error: {e}")
```
